## 1) UCI 데이터셋 불러오기 (URL을 통해)

홈페이지에 있는 여러 데이터셋 중 (`Adult`처럼 강의때 했던 내용을 제외하고) 

우측에 있는 "Most Popular" 섹션 가운데에서 한 데이터셋을 선택한 뒤, 해당 데이터셋을 URL로 부터 바로 읽어보세요. (작업 코드를 작성하며, 아래의 질문에 대해 대답할 수 있어야합니다.) 

- 헤더가 정상적으로 보여지나요?
- 첫 5줄 혹은 마지막 5줄을 확인하고 각각 순서를 가지고 정렬이 되어있나요?
- UCI 홈페이지에서 설명한 것과 동일한 행과 열을 가지고 있나요?
- 주의 할 점으로, UCI는 y 변수 ( 나중에 머신러닝 모델링에서 쓰기 위한 값으로 지금 당장은 중요 하지 않습니다)를 "속성"이 아니라, "클래스 속성"으로 취급하기 때문에 행이 1개 더 많은 경우도 있습니다. 
- 데이터셋에 결측치가 존재 하나요? (UCI의 설명과 일치 하는지)
- `NaN` 혹은 다른 결측치를 다루는 작업을 `df.isnull().sum()`외의 방법으로 할 수 있나요? (예를 들면, `np.NaN`으로 표기된 값을 결측치로 처리해야 합니다) 
- `.describe()` 함수를 사용하여, 여러 통계치를 확인할 수 있나요?

```python
# 1) UCI 데이터셋 불러오기 (URL을 통해)
data_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/zoo/zoo.data'

import pandas as pd

df1 = pd.read_csv(data_url)
df1.columns = ['animal name', 'hair', 'feathers', 'eggs', 'milk', 'airborne', 'aquatic', 
              'predator', 'toothed', 'backbone', 'breathes', 'venomous', 'fins', 'legs', 'tail', 'domestic', 'catsize', 'type']
print(df1)

# 헤더가 정상적으로 보여지나요?
print(list(df1.columns.values))
print('---')
# 첫 5줄 혹은 마지막 5줄을 확인하고 각각 순서를 가지고 정렬이 되어있나요?
  # 네
print(df1.head())
print('---')
print(df1.tail())
print('---')

# UCI 홈페이지에서 설명한 것과 동일한 행과 열을 가지고 있나요?
  # UCI says   Number of Instances: 101
  #            Number of Attributes: 17
  # df.shape = (100, 18)가 나옵니다.
print(df1.shape)
print('---')

# 데이터셋에 결측치가 존재 하나요? (UCI의 설명과 일치 하는지)
  # 아니오 (UCI says : Missing Values? No) 
  # UCI 설명과 일치합니다.
print(df1.isnull().all())
print('---')

# NaN 혹은 다른 결측치를 다루는 작업을 df.isnull().sum()외의 방법으로 할 수 있나요? (예를 들면, np.NaN으로 표기된 값을 결측치로 처리해야 합니다)
## 결측치가 존재한다는 가정 하 작성하였습니다.
print(df1.fillna(0)) # 모든 NaN을 0으로 치환
print('---')
print(df1.dropna()) # NaN을 하나라도 포함하는 행 삭제(default : how='any', axis=0)
print('---')
print(df1.dropna(how='all')) # 행 전체가 NaN인 행만 삭제
print('---')

# .describe() 함수를 사용하여, 여러 통계치를 확인할 수 있나요?
print(df1.describe())
```
